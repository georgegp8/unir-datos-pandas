# Fusion de varios Dataframes

```python
import pandas as pd
```

## Sección 1: Total de usuarios en un mes

Tu objetivo es encontrar el número total de viajes en la estación Wilson en días laborables de julio.

```python
cal = pd.read_csv('../cta_calendar.csv')
ridership = pd.read_csv('../cta_ridership.csv')
stations = pd.read_pickle('../stations.p')
```

### Instrucciones 1/3
Fusiona las tablas ridership y cal en las columnas year, month, day.

```python
ridership_cal = ridership.merge(cal, on=['year', 'month', 'day'])
print(ridership_cal.head())
```

### Instrucciones 2/3
Amplía la fusión anterior a tres tablas fusionando también la tabla stations.

```python
ridership_cal_stations = ridership.merge(cal, on=['year', 'month', 'day']).merge(stations, on='station_id')
print(ridership_cal_stations.head())
```

### Instrucciones 3/3
Crea un filtro para julio, días laborables, estación Wilson y suma el total de viajes.

```python
filter_criteria = ((ridership_cal_stations['month'] == 7) & (ridership_cal_stations['day_type'] == 'Weekday') & (ridership_cal_stations['station_name'] == 'Wilson'))
total_rides = ridership_cal_stations.loc[filter_criteria, 'rides'].sum()
print(f'Total de viajes: {total_rides}')
```

## Sección 2: Fusión de tres tablas

Fusionarás las tablas licenses y wards con la tabla zip_demo para analizar la renta media por código postal.

```python
licenses = pd.read_csv('../licenses.csv')
wards = pd.read_csv('../Chicago_wards.csv')
zip_demo = pd.read_csv('../zip_demo.csv')
```

### Instrucciones 1/2
Partiendo de licenses, fusiona con zip_demo en zip y luego con wards en ward.

```python
licenses_zip_ward = licenses.merge(zip_demo, on='zip').merge(wards, on='ward')
print(licenses_zip_ward.head())
```

### Instrucciones 2/2
Agrupa por alderman y calcula la mediana de income.

```python
result = licenses_zip_ward.groupby('alderman').agg({'income': 'median'})
print(result)
```

## Sección 3: Fusión de uno a muchos con varias tablas

Unirás las tablas land_use, census y licenses para analizar cómo se relacionan población, lotes vacantes y número de cuentas.

```python
land_use = pd.read_csv('../land_use.csv')
census = pd.read_csv('../Chicago_census.csv')
licenses = pd.read_csv('../licenses.csv')
```

### Instrucciones 1/3
Fusiona land_use con census y luego con licenses usando sufijos _cen y _lic.

```python
land_cen_lic = land_use.merge(census, on='ward').merge(licenses, on='ward', suffixes=('_cen', '_lic'))
print(land_cen_lic.head())
```

### Instrucciones 2/3
Agrupa por ward, pop_2010 y vacant. Luego cuenta el número de accounts.

```python
pop_vac_lic = land_cen_lic.groupby(['ward', 'pop_2010', 'vacant'], as_index=False).agg({'account': 'count'})
print(pop_vac_lic.head())
```

### Instrucciones 3/3
Ordena pop_vac_lic por vacant (desc), account (asc) y pop_2010 (asc).

```python
sorted_pop_vac_lic = pop_vac_lic.sort_values(['vacant', 'account', 'pop_2010'], ascending=[False, True, True])
print(sorted_pop_vac_lic.head())
```
