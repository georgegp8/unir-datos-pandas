# Fusionar Tabla Consigo Misma

## Seccion 1: Autounion

Fusionar una tabla consigo misma puede ser util cuando quieras comparar valores de una columna con otros valores de la misma columna.

En este ejercicio, practicaras creando una tabla en la que, para cada pelicula, figuraran en una fila el director de la pelicula y un miembro del equipo.

Te han dado una tabla llamada `crews`, que tiene las columnas `id`, `job` y `name`.

### Instruccion 1/3
En una variable llamada `crews_self_merged`, une la tabla `crews` a si misma en la columna `id` mediante una union interna, estableciendo los sufijos en `'_dir'` y `'_crew'` para las tablas izquierda y derecha respectivamente.

```python
import pandas as pd

# Carga de la tabla crews
crews = pd.read_pickle('../crews.p')

# Instruccion 1/3
crews_self_merged = crews.merge(crews, on='id', how='inner', suffixes=('_dir', '_crew'))
```

### Instruccion 2/3
Crea un indice booleano, llamado `boolean_filter`, que seleccione filas de la tabla izquierda con el job de `'Director'` y evita las filas con el job `'Director'` en la tabla derecha.

```python
# Instruccion 2/3
boolean_filter = ((crews_self_merged['job_dir'] == 'Director') &
                  (crews_self_merged['job_crew'] != 'Director'))
direct_crews = crews_self_merged[boolean_filter]
```
