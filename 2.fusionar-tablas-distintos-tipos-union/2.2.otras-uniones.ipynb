# Otras Uniones

## Seccion 1: Realiza union a la derecha para encontrar peliculas unicas

La mayoria de las peliculas recientes de ciencia ficcion de gran presupuesto tambien pueden clasificarse como peliculas de accion.

Te dan una tabla de peliculas de ciencia ficcion llamada `scifi_movies` y otra tabla de peliculas de accion llamada `action_movies`. Tu objetivo es encontrar que peliculas se consideran solo de ciencia ficcion.

Una vez que tengas esta tabla, puedes fusionar la tabla `movies` para ver los nombres de las peliculas.

### Instruccion 1/4
Fusiona las tablas `action_movies` y `scifi_movies` con una union a la derecha en `movie_id`. Guarda el resultado como `action_scifi`.

```python
import pandas as pd

# Carga de datos base
movies = pd.read_csv('../movies.csv')
movie_to_genres = pd.read_csv('../movie_to_genres.csv')

# Tablas del ejercicio
scifi_movies = movie_to_genres[movie_to_genres['genre'] == 'Science Fiction'][['movie_id', 'genre']]
action_movies = movie_to_genres[movie_to_genres['genre'] == 'Action'][['movie_id', 'genre']]

# Instruccion 1/4
action_scifi = action_movies.merge(scifi_movies, how='right', left_on='movie_id', right_on='movie_id')
```

### Instruccion 2/4
Actualiza la fusion para anadir sufijos, donde `'_act'` y `'_sci'` son sufijos para las tablas izquierda y derecha, respectivamente.

```python
# Instruccion 2/4
action_scifi = action_movies.merge(scifi_movies, on='movie_id', how='right', suffixes=('_act', '_sci'))
print(action_scifi.head())
```

### Instruccion 3/4
De `action_scifi`, subconjunta solo las filas en las que la columna `genre_act` sea nula.

```python
# Instruccion 3/4
scifi_only = action_scifi[action_scifi['genre_act'].isnull()]
```

### Instruccion 4/4
Fusiona `movies` y `scifi_only` utilizando la columna `id` de la tabla izquierda y la columna `movie_id` de la tabla derecha con una union interna. Luego imprime las primeras filas y la forma del resultado.

```python
# Instruccion 4/4
movies_and_scifi_only = movies.merge(scifi_only, left_on='id', right_on='movie_id')
print(movies_and_scifi_only.head())
print(movies_and_scifi_only.shape)
```
