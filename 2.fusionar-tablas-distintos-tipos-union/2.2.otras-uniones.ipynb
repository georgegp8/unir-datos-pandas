# Otras Uniones

## Seccion 1: Realiza union a la derecha para encontrar peliculas unicas

La mayoria de las peliculas recientes de ciencia ficcion de gran presupuesto tambien pueden clasificarse como peliculas de accion.

Te dan una tabla de peliculas de ciencia ficcion llamada `scifi_movies` y otra tabla de peliculas de accion llamada `action_movies`. Tu objetivo es encontrar que peliculas se consideran solo de ciencia ficcion.

Una vez que tengas esta tabla, puedes fusionar la tabla `movies` para ver los nombres de las peliculas.

```python
import pandas as pd

# Carga de datos base
movies = pd.read_csv('../movies.csv')
movie_to_genres = pd.read_csv('../movie_to_genres.csv')

# Tablas del ejercicio
scifi_movies = movie_to_genres[movie_to_genres['genre'] == 'Science Fiction'][['movie_id', 'genre']]
action_movies = movie_to_genres[movie_to_genres['genre'] == 'Action'][['movie_id', 'genre']]
```

### Instrucciones 1/4
Fusiona las tablas `action_movies` y `scifi_movies` con una union a la derecha en `movie_id`. Guarda el resultado como `action_scifi`.

```python
action_scifi = action_movies.merge(scifi_movies, how='right', left_on='movie_id', right_on='movie_id')
```

### Instrucciones 2/4
Actualiza la fusion para anadir sufijos, donde `'_act'` y `'_sci'` son sufijos para las tablas izquierda y derecha, respectivamente.

```python
action_scifi = action_movies.merge(scifi_movies, on='movie_id', how='right', suffixes=('_act', '_sci'))
print(action_scifi.head())
```

### Instrucciones 3/4
De `action_scifi`, subconjunta solo las filas en las que la columna `genre_act` sea nula.

```python
scifi_only = action_scifi[action_scifi['genre_act'].isnull()]
```

### Instrucciones 4/4
Fusiona `movies` y `scifi_only` utilizando la columna `id` de la tabla izquierda y la columna `movie_id` de la tabla derecha con una union interna. Luego imprime las primeras filas y la forma del resultado.

```python
movies_and_scifi_only = movies.merge(scifi_only, left_on='id', right_on='movie_id')
print(movies_and_scifi_only.head())
print(movies_and_scifi_only.shape)
```

## Seccion 2: Generos populares con union a la derecha

¿Cuales son los generos de las peliculas mas populares? Para responder a esta pregunta, tienes que fusionar los datos de las tablas `movies` y `movie_to_genres`. En una tabla llamada `pop_movies`, se han seleccionado las 10 peliculas mas populares de la tabla `movies`. Para asegurarte de que estas analizando todas las peliculas populares, fusionala con la tabla `movie_to_genres` utilizando una union a la derecha. Para completar tu analisis, cuenta el numero de generos diferentes.

Las tablas `pop_movies` y `movie_to_genres` se han cargado para ti.

```python
import matplotlib.pyplot as plt

# Tabla del ejercicio: 10 peliculas mas populares
pop_movies = movies.nlargest(10, 'popularity')
```

### Instrucciones 1/3
Fusiona `movie_to_genres` y `pop_movies` utilizando una union a la derecha. Guarda los resultados como `genres_movies`.

```python
genres_movies = movie_to_genres.merge(pop_movies, how='right',
                                       left_on='movie_id',
                                       right_on='id')
```

### Instrucciones 2/3
Agrupa `genres_movies` por `genre` y cuenta el numero de valores de `id`.

```python
genre_count = genres_movies.groupby('genre').agg({'id': 'count'})
```

### Instrucciones 3/3
Grafica `genre_count` con un grafico de barras y muestralo.

```python
genre_count.plot(kind='bar')
plt.show()
```

## Seccion 3: Utilizar union externa para seleccionar actores

Un aspecto interesante de utilizar una union externa es que, como devuelve todas las filas de ambas tablas fusionadas y nulas cuando no coinciden, puedes utilizarla para encontrar filas que no tienen coincidencia en la otra tabla. Para que lo pruebes por ti mismo, te han dado dos tablas con una lista de actores de dos peliculas populares: *Iron Man 1* e *Iron Man 2*. La mayoria de los actores actuaron en ambas peliculas. Utiliza una union externa para encontrar actores que **no** actuaron en ambas peliculas.

La tabla *Iron Man 1* se llama `iron_1_actors` y la tabla *Iron Man 2* se llama `iron_2_actors`. Se han cargado ambas tablas y se han impreso algunas filas para que puedas ver la estructura.

```python
# Tablas del ejercicio: actores de Iron Man 1 e Iron Man 2
iron_1_actors = pd.read_csv('../iron_1_actors.csv')
iron_2_actors = pd.read_csv('../iron_2_actors.csv')

print(iron_1_actors)
print(iron_2_actors)
```

### Instrucciones 1/2
Guarda en `iron_1_and_2` la fusion de las tablas `iron_1_actors` (izquierda) con `iron_2_actors` con una union externa en la columna `id` y establece los sufijos en `('_1','_2')`.

```python
iron_1_and_2 = iron_1_actors.merge(iron_2_actors,
                                    on='id',
                                    how='outer',
                                    suffixes=('_1', '_2'))
```

### Instrucciones 2/2
Crea un indice que devuelva `True` si `name_1` o `name_2` son nulos y `False` en caso contrario.

```python
m = ((iron_1_and_2['name_1'].isnull()) |
     (iron_1_and_2['name_2'].isnull()))

print(iron_1_and_2[m].head())
```
