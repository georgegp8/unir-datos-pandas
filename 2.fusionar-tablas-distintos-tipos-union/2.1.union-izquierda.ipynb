# Fusionar Tablas con Distintos Tipos de Union

## Seccion 1: Contar las filas que faltan con la union a la izquierda

Hay voluntarios que recogen datos y los introducen en la base de datos de peliculas The Movie Database (TMDB). Esto incluye datos financieros, como el presupuesto de la pelicula y los ingresos. Si quisieras saber a que peliculas les faltan datos, podrias utilizar una union a la izquierda para identificarlas.

Las tablas `movies` y `financials` se han cargado para ti.

### Instruccion 1/3
¿Que columna es probablemente la mejor para fusionar las dos tablas?

Respuestas posibles:
- `on='budget'`
- `on='popularity'`
- `on='id'`

```python
# pyright: reportMissingModuleSource=false
import pandas as pd

# Carga de tablas
movies = pd.read_csv('../movies.csv')
financials = pd.read_csv('../financials.csv')

# Instruccion 1/3: la mejor columna para fusionar es id
merge_key = 'id'
print(merge_key)
```

### Instruccion 2/3
Fusiona la tabla `movies` como tabla izquierda, con la tabla `financials` mediante una union a la izquierda y guarda el resultado en `movies_financials`.

```python
# Instruccion 2/3: left join usando la clave correcta
movies_financials = movies.merge(financials, on='id', how='left')
print(movies_financials.head())
```

### Instruccion 3/3
Cuenta el numero de filas de `movies_financials` con un valor nulo en la columna `budget`.

```python
# Instruccion 3/3: contar filas con budget nulo
number_of_missing_fin = movies_financials['budget'].isna().sum()
print(number_of_missing_fin)
```

## Seccion 2: Enriquecer un conjunto de datos

Establecer `how='left'` con el metodo `.merge()` es una tecnica util para enriquecer un conjunto de datos con informacion adicional de una tabla diferente.

En este ejercicio, empezarás con una muestra de datos de peliculas de la serie *Toy Story*. Tu objetivo es enriquecer estos datos añadiendo el eslogan de marketing de cada pelicula. Compararas los resultados de una union a la izquierda frente a los de una union interna.

### Instruccion 1/2
Fusiona `toy_story` y `taglines` en la columna `id` con una union a la izquierda, guarda el resultado como `toystory_tag`, luego imprime `toystory_tag` y su `shape`.

### Instruccion 2/2
Con `toy_story` como tabla izquierda, fusiona `taglines` en la columna `id` con una union interna, guarda el resultado como `toystory_tag`, luego imprime `toystory_tag` y su `shape`.

```python
# Preparacion de tablas para la seccion 2
movies = pd.read_csv('../movies.csv')
taglines = pd.read_pickle('../taglines.p')
toy_story = movies.loc[movies['title'].str.contains('Toy Story', na=False), ['id', 'title']]

# Instruccion 1/2: left join
toystory_tag = toy_story.merge(taglines, on='id', how='left')
print(toystory_tag)
print(toystory_tag.shape)
```

```python
# Instruccion 2/2: inner join
toystory_tag = toy_story.merge(taglines, on='id')
print(toystory_tag)
print(toystory_tag.shape)
```
