# Fusión de indices para clasificar películas

Para practicar la fusión en indices, fusionarás `movies` y una tabla llamada `ratings` que contiene información sobre la clasificación de películas. Asegúrate de que tu fusión devuelva **todas** las filas de la tabla `movies` y solo los filas coincidentes de la tabla `ratings`.

Las tablas `movies` y `ratings` se han cargado para ti.

```python
import pandas as pd

# Carga de datos base
movies = pd.read_csv('../movies.csv')
ratings = pd.read_csv('../ratings.csv')
```

### Instrucciones 1/1
Fusiona las tablas `movies` y `ratings` en la columna `id`, manteniendo todas las filas de la tabla `movies`, y guardo el resultado como `movies_ratings`.

```python
movies_ratings = movies.merge(ratings, on='id', how='left')

print(movies_ratings.head())
```
